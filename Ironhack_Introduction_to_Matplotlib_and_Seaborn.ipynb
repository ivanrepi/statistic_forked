## Matplotlib

### Introducción

Matplotlib es una librería de visualización de datos basada en arrays de numpy. Esta librería tiene mucho recorrido, ya que su primera versión 0.1 está datada en 2003. La mayor ventaja de esta librería es que nos permite visualizar datos en Python de una forma muy flexible, independientemente de la versión y el sistema operativo.

### Primeros pasos

Desde que empezamos a trabajar con Jupyter Notebook siempre comenzamos introduciendo la siguiente linea de código:

```python
%matplotlib inline
```

Esto va a asegurarnos que todos los gráficos que creemos sean mostrados dentro del notebook.

Posteriormente importamos matplotlib, y pyplot desde la librería de matplotlib:

```python
import matplotlib
import matplotlib.pyplot as plt
```

### Gráficos Básicos

#### Gráficos de lineas

Podemos crear dos arrays utilizando numpy y dibujar un gráfico de líneas:

```python
import numpy as np
```

```python
var1 = np.arange(0,10,1)
var2 = var1**2

plt.plot(var1,var2)
```

Aquí estamos utilizando la función arange de numpy para generar números enteros de 0 a 9. Luego dibujamos esos números y estos mismos al cuadrado.

#### Gráficos de dispersión
Aquí usamos la función scatter para trazar los mismos dos arrays

```python
plt.scatter(var1, var2)
```

Este tipo de gráficos se utilizan, como su propio nombre indica, para comprobar la dispersión existente en la relación entre dos variables. En el caso anterior observamos una relación cuadrática perfecta ya que una array es dependiente del otro, pero en datos reales no va a suceder esto. Vamos a simular dos variables aleatorias para ver cómo sería un gráfico de dispersión normal

```python
from numpy import random
```

```python
var3 = random.normal(0,1,100)
var4 = random.normal(0,1,100)

plt.scatter(var3,var4)
```

En este caso vemos como, simulando dos variables normales con media 0 y desviación típica 1, el gráfico de dispersión obtenido se parece más a un caso real.

Sin embargo, si utilizamos un gráico de lineas para dibujar esta relación, podemos ver que no tiene tanto sentido:

```python
plt.plot(var3,var4)
```

## Seaborn

### Introducción
`Seaborn` es una biblioteca creada para complementar `matplotlib`. Su objetivo es optimizar la cantidad de código necesario para crear visualizaciones con `matplotlib`. También nos permite trabajar con los datasets de pandas. Esta es la principal desventaja que resuelve `Seaborn` respecto a `matplotlib`.

### Primeros pasos

Después de configurar `%matplotlib inline` e importar `matplotlib`, importamos `seaborn` como `sns`. También podemos establecer un estilo por defecto en seaborn.

```python
import seaborn as sns
sns.set()
```

### Gráficos básicos

#### Gráficos de lineas

Podemos generar un dataframe con pandas utilizando las dos variables que hemos creado antes:

```python
import pandas as pd
```

```python
df1 = pd.DataFrame()
df1['var1'] = var1
df1['var2'] = var2

df1.head()
```

Ahora que ya tenemos nuestro dataframe creado ya podemos dibujar el gráfico de líneas con `seaborn`

```python
sns.lineplot(df1.var1, df1.var2)
```

#### Gráficos de dispersión

Utilizamos la función `lmplot` para dibujar un gráfico de dispersión, que se llama así porque es el gráfico típico para estudiar la relación entre dos variables en una regresión linear o linear regression (lm). En este caso vamos a fijar el parámetro `fit_reg = False` ya que no estamos interesados en estudiar esto en concreto.

```python
sns.lmplot("var1", "var2", data=df1, fit_reg=True)
```

Vamos a ver el mismo ejemplo que antes, ajustando en este caso la recta de regresión para entender qué significa

```python
df2 = pd.DataFrame()
df2['var3'] = var3
df2['var4'] = var4

df2.head()
```

```python
sns.lineplot(df1.var1, df2.var3)
```

```python
sns.lmplot("var3", "var4", data=df2, fit_reg=True)
```

## Utilizando Pandas directamente

La librería `matplotlib` está integrada en `pandas`, por lo que podremos dibujar los mismos gráficos utilizando directamente la librería `pandas`. Es importante recordar que en este caso también necesitamos ajustar los gráficos para que sean dibujados en Jupyter Notebook con el comando `%matplotlib inline`

### Gráficos básicos

Accedemos a la librería de `matplotlib` desde `pandas` llamando a la función `.plot`. Posteriormente especificamos el gráfico que queremos utilizar con sus parámetros.

Vamos a ver como funciona:

#### Gráficos de lineas:

```python
df1.plot.line(x="var1", y="var2")
```

#### Gráfico de dispersión:

```python
df2.plot.scatter(x="var3", y="var4",c="blue")
```

#### Histograma

Otro gráfico muy informativo y del que ya hemos hablado previamente es el histograma. Dibujar un histograma es muy sencillo con cualquiera de estos métodos. Utilizando `matplotlib` a través de `pandas` podríamos hacerlo de la siguiente forma:

```python
df2.plot.hist(x="var3")
```

#### Gráfico de Caja

También podremos observar una información similar a la que visualizamos en el histograma con un boxplot. Este tipo de gráfico podemos obtenerlo de la siguiente forma:

```python
df2.plot.box(x="var3")
```
