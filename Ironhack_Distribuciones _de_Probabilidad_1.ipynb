# Distribuciones de Probabilidad

## Introducción

La distribución de probabilidad, se refiere a todos los resultados posibles que pueda tener una variable aleatoria, es decir, describe el comportamiento de dicha variable dentro de un intervalo de valores o de posibles resultados.

La variable aleatoria puede ser discreta o continua. Una variable aleatoria discreta es aquella representada por números enteros, caracterizada por el límite de valores que puede tomar. Por otro lado, una variable aleatoria continua no posee esta separación o limitación, puede tomar cualquier valor dentro del límite establecido.

Una variable aleatoria es una variable cuyos valores son el resultado de un experimento aleatorio. Cada resultado posible en una variable aleatoria tiene una cierta probabilidad derivada de la distribución de probabilidad.

## Distribuciones de Probabilidad Discreta

Una distribución de probabilidad discreta se caracteriza por tener un número finito o infinitamente contable de resultados en el espacio muestral. La suma de probabilidades para todos los resultados en el espacio muestral debe sumar 1.

### Distribución de probabilidad binomial

Es una distribución de probabilidad de una variable aleatoria discreta que resulta de contar el número de éxitos al repetir n veces (n ensayos de bernouilli) un experimento que tiene dos resultados posibles (éxito y fracaso) con probabilidades p y q respectivamente.



3. Ejemplo distribución binomial

En un hospital se ha registrado una probabilidad del 55% de que un evento ocurra cada vez que se ingresa a una persona en él. ¿Cuál es la probabilidad de que ocurra en 2 de 5 personas que han ingresado hoy?

```python
from scipy.special import comb

#número de ensayos (n) = 5
#número de éxitos (k) = 2
#probabilidad éxito (p) = 0.55
#probabilidad fracaso (q) = 0.45

def calculate_binomial_function(p,q,n,k):
    return comb(n, k) * p ** k * q **(n - k)
#B(5, 0.55)
prob = calculate_binomial_function(0.55, 0.45, 5, 2)
print('La probabilidad de que ocurra en 2 de las 5 personas que han ingresado hoy es de %.4f' % prob)
```

### Distribución de Poisson

Es una distribución de probabilidad de una variable aleatoria discreta que nos proporciona la probabilidad de que ocurra un determinado evento un número de veces K en un intervalo determinado de tiempo, longitud, área...
                       


Mu es el número medio de veces que ocurre el suceso en el intervalo de tiempo, longitud, área...

La distribución de Poisson se puede utilizar como una aproximación a la binomial cuando el número de ensayos independientes es grande y la probabilidad de éxito sea pequeña.

```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

1. Ejemplo cálculo función de distribución de poisson

En un hospital de una ciudad importante se están estudiando los nacimientos de varones. Se sabe que en una semana nacen una media de 7 varones. Calcula:

- Probabilidad de qure nazcan 3 varones en una semana
- Probabilidad de que nazcan menos de 3 varones en una semana

```python
# p(x=3)
def calculate_poisson_function (mu, k):
    return (((mu ** k) * (np.e ** (- mu))) / math.factorial(k))*100

prob_porc = calculate_poisson_function(7, 3)
print ('La probabilidad de que nazcan 3 varones esta semana es del %.2f por ciento' % prob_porc)
```

```python
# p(x<3) = p(x=0) + p(x=1) + p(x=2)

prob_porc = calculate_poisson_function(7,0) + calculate_poisson_function(7,1) + calculate_poisson_function(7,2)

print ('La probabilidad de que nazcan menos de 3 varones esta semana es del %.2f por ciento' % prob_porc)
```

### Distribución Geométrica

Es una distribución de probabilidad de una variable aleatoria discreta que nos proporciona la probabilidad de tener que esperar exactamente x pruebas hasta encontrar el primer éxito si la probabilidad de éxito en una sola prueba es p. Por ejemplo, en un proceso de selección, podría definir el número de entrevistas que deberíamos realizar antes de encontrar al primer candidato aceptable.

Para que se dé esta distribución se cumple que:

- Todos los n intentos repetidos son independientes
- La probabilidad de éxito permanece constante en todos los intentos

p = probabilidad de éxito
q = 1-p = probabilidad de fracaso

                                                p(x,p) = p(1−p) ^ (x−1)

4. Ejemplo distribución geométrica. 

Si la probabilidad de que un tirador experto dé en el blanco es del 95%. ¿Cuál es la probabilidad de que falle por primera vez en su decimoquinto disparo?

```python
#Ojo. En este caso nuestra probabilidad de éxito es la probabilidad de que falle el tirador.

def calculate_geom_function(x, p):
    return  p * (1-p) ** (x - 1)

prob = calculate_geom_function(15, 0.05)
print ('La probabilidad de que falle el tirador por primera vez en su decimoquinto disparo es de %.4f' % prob)
```

## Teorema del límite central

El teorema describe la distribución de la media de una muestra aleatoria proveniente de una población con varianza finita. Cuando el tamaño de la muestra es lo suficientemente grande, la distribución de las medias sigue aproximadamente una distribución normal. El teorema se aplica independientemente de la forma de la distribución de la población. Muchos procedimientos estadísticos comunes requieren que los datos sean aproximadamente normales. El teorema de límite central le permite aplicar estos procedimientos útiles a poblaciones que son considerablemente no normales. El tamaño que debe tener la muestra depende de la forma de la distribución original. 

Si la distribución de la población es simétrica, un tamaño de muestra de 5 podría producir una aproximación adecuada. Si la distribución de la población es considerablemente asimétrica, es necesario un tamaño de muestra más grande. Por ejemplo, la distribución de la media puede ser aproximadamente normal si el tamaño de la muestra es mayor que 50. Las siguientes gráficas muestran ejemplos de cómo la distribución afecta el tamaño de la muestra que se necesita.

```python
# Ejemplo teorema del límite central
muestra_binomial = []
muestra_exp = []
muestra_possion = []
muestra_geometric = []
mu = 0.9
lam = 1.0
size=1000

for i in range(1,20000):
    muestra = np.random.binomial(1, mu, size=size)
    muestra_binomial.append(muestra.mean())
    muestra = np.random.exponential(scale=2.0,size=size)
    muestra_exp.append(muestra.mean())
    muestra = np.random.geometric(p=.5, size=size)
    muestra_geometric.append(muestra.mean())
    muestra = np.random.poisson (lam=lam, size=size)
    muestra_possion.append(muestra.mean()) 

df = pd.DataFrame({ 'binomial' : muestra_binomial, 
                     'poission' : muestra_possion,
                     'geometrica' : muestra_geometric,
                    'exponencial' : muestra_exp})

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
df.binomial.hist(ax=axes[0,0], alpha=0.9, bins=1000)
df.exponencial.hist(ax=axes[0,1],bins=1000)
df.poission.hist(ax=axes[1,0],bins=1000)
df.geometrica.hist(ax=axes[1,1],bins=1000)

axes[0,0].set_title('Binomial')
axes[0,1].set_title('Poisson')
axes[1,0].set_title('Geométrica')
axes[1,1].set_title('Exponencial')
plt.show()
```
