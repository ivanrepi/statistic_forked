# Mejorando la estética de nuestros gráficos

## Introducción
Si bien es posible utilizar el modo predeterminado para todas las visualizaciones y seguir transmitiendo correctamente el mensaje, la personalización de las visualizaciones hace que el mensaje sea más claro y los gráficos más atractivos desde el punto de vista visual.

## Customización de los gráficos de barras
### Añadiendo títulos, leyendas y cambiando el tamaño
La adición de un título se puede hacer pasando un argumento title a la función del gráfico. Se puede añadir una leyenda estableciendo el argumento de la leyenda como verdadero. También podemos redimensionar pasando una anchura y una altura al argumento figsize. Por ejemplo, utilizando el conjunto de datos de vehículos de lecciones anteriores podemos hacer los siguiente:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vehicles = pd.read_csv("./vehicles.csv")
```

```python
vehicles['Drivetrain'].value_counts().plot(kind="bar", title="Count of Vehicles by Type of Drivetrain", figsize=(10, 2), legend=True)
```

### Colores
Al trazar un gráfico de barras, podemos usar el esquema de color estándar, pero también podemos usar otro esquema de color.

Matplotlib tiene una larga lista de mapas de color preestablecidos:



El problema principal es que estos mapas de color tienen una longitud establecida que a veces es mayor que la de nuestros datos. Para asegurarnos de que obtenemos una muestra diversa del mapa de color, debemos obtener el número correcto de colores del mapa de color.

Hacemos esto primero averiguando el número de valores únicos en nuestra columna:

```python
vehicles.Drivetrain.nunique()
```

Luego generamos una matriz de valores de 0 a 1 menos el número que se corresponde a la cantidad de valores únicos en la columna

```python
np.arange(vehicles.Drivetrain.nunique())
```

Seleccionamos un mapa de colores y le aplicamos la longitud. En este caso elegimos el mapa de color emparejado.


```python
plt.cm.Paired(np.arange(vehicles.Drivetrain.nunique()))
```

Finalmente, trazamos nuestro gráfico de barras:

```python
vehicles['Drivetrain'].value_counts().plot(kind="bar",
                                           color=plt.cm.Dark2(np.arange(vehicles.Drivetrain.nunique())));
```

## Customizando gráficos de dispersión
### Cambiando el alpha
El canal alfa mide la opacidad. Alpha varía en el intervalo[0, 1], donde el valor 0 corresponde a transparencia total, y 1 a opacidad. A veces los puntos tienden a agruparse. La reducción del alfa (es decir, el aumento de la transparencia) nos ayuda a detectar claramente puntos distintos incluso cuando hay un grupo denso de puntos.

Observa que en el gráfico de abajo es difícil saber cuántos puntos se superponen. Sin embargo, en el siguiente gráfico con alfa=0.2, es más fácil de decir.

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year");
```

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year", alpha=0.2);
```

### Añadiendo una cuadrícula
Tener una cuadrícula puede ayudar a leer una gráfica de dispersión. Para añadir una cuadrícula, establecemos el argumento `grid` como verdadero.

```python
vehicles.plot.scatter(x="CO2 Emission Grams/Mile", y="Fuel Barrels/Year", alpha=0.2, grid=True);
```
