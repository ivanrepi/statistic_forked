## Introducción
En la lección de Introducción a la Probabilidad, introdujimos la probabilidad condicional, que es un prerrequisito para la Estadística Bayesiana.



Una probabilidad condicional es una probabilidad basada en cierta información de fondo. Su notación se expresa típicamente como P(A|B), lo que significa la probabilidad de que ocurra el evento A dado que alguna condición B es cierta. Vimos que esto podía ser calculado usando la siguiente ecuación.

$P(A|B) = \frac{P(A∩B)}{P(B)}$

En otras palabras, la probabilidad de A dado B es igual a la probabilidad de la intersección de A y B dividida por la probabilidad de B.

También mencionamos brevemente la independencia y dependencia de los acontecimientos y dijimos que cuando los acontecimientos son independientes, al menos una de las siguientes condiciones debe ser cierta:

$P(A|B) = P(A)$

$P(B|A) = P(B)$

$P(A∩B) = P(A) P(B)$

En otras palabras, si los eventos son verdaderamente independientes, entonces la probabilidad de uno dado a otro debería ser simplemente la probabilidad de que ocurra ese evento. De manera similar, la probabilidad de la intersección de los dos eventos debe ser simplemente el producto de las probabilidades de los dos eventos, ya que ninguno de los dos impacta al otro.

## Enfoque Bayesiano vs. Enfoque Frequentista

Para un bayesiano convencido, un p-valor o un t-ratio es una herejía, ya que lo único que necesitamos para contrastar hipótesis es tener la distribución a posteriori de los parámetros.

un bayesiano proporcionaría conclusiones del tipo: «hay un 95% de probabilidad de que el parámetro esté entre 2,6 y 5,5”; sin embargo, un frecuentista afirmaría: “si generáramos 100 muestras aleatorias del mismo tamaño y repitiéramos la estimación 100 veces, aproximadamente en 95 de ellas el parámetro estimado estaría entre 2,6 y 5,5”.

En el enfoque bayesiano se combinan evidencias subjetivamente acumuladas con la información objetiva de los datos. Por eso la probabilidad bayesiana se llama probabilidad subjetiva o condicional: contiene información sobre observaciones y experiencias previas a diferencia de la probabilidad frecuentista, que se basa solo en el análisis de las frecuencias de los datos y no admite suposiciones a priori de ninguna clase aparte de las hipótesis.

De esta forma, mientras los métodos estadísticos tradicionales se preguntan qué nos dicen los datos acerca del valor del parámetro (ignorando con ello toda evidencia externa a los datos), los métodos bayesianos se preguntan explícitamente cómo cambia nuestro estado de información acerca del valor del parámetro (o de cualquier otra cantidad de interés) a la luz de los datos observados.

Si bien esto es cierto, conforme se va acumulando la evidencia, las opiniones de esas dos personas deberían converger.

Al admitir información adicional, los métodos bayesianos reconocen que cada problema es distinto y promueven que el procedimiento de análisis se adapte al problema en cuestión, y no al revés; como consecuencia, tienden a ser más flexibles. Sin embargo, para poder hacer esto posible, la implementación de las técnicas bayesianas usualmente requiere de un esfuerzo computacional muy alto.

Afortunadamente, durante la segunda mitad del siglo XX se comenzaron a desarrollar técnicas numéricas flexibles y eficientes basadas en métodos de simulación estocástica. Esto, aunado al desarrollo de la tecnología que ha dado lugar a una mayor capacidad de procesamiento y de almacenamiento de los equipos de cómputo gracias a la virtualización, el Cloud Computing y el Big Data, está provocando un auge de los métodos bayesianos.

## Teorema de Bayes 
Para eventos dependientes, necesitamos modificar el último conjunto de ecuaciones para tener en cuenta el hecho de que un evento impacta al otro.

$P(A∩B) = P(A) P(B|A)$

$P(B∩A) = P(B) P(A|B)$

Así que la intersección de A y B es igual a la probabilidad de A multiplicada por la probabilidad de B dada A y, puesto que no hay nada especial en A frente a B, también es igual a la probabilidad de B multiplicada por la probabilidad de A dada B.

Si este es el caso, entonces:

$P(B) P(A|B) = P(A) P(B|A)$

Y si dividimos ambos lados por P(B), terminamos con lo siguiente.

$P(A|B) = \frac{P(A) P(B|A)}{P(B)}$

Esta ecuación se conoce como Teorema de Bayes, y es una ecuación sorprendentemente poderosa que forma la idea fundamental detrás de todas las estadísticas bayesianas. Nos permite establecer lo que creemos sobre la probabilidad de una hipótesis antes de ver cualquier dato, y también nos proporciona un medio para actualizar esas creencias una vez que hayamos visto algunos datos.

Esta ecuación tiene algunos componentes diferentes, así que vamos a descomponerlos:

* $P(A)$ se conoce como *probabilidad previa* o *a priori*, y es la probabilidad de la hipótesis antes de que veamos los datos.
* $P(B|A)$ se conoce como *verosimilitud*, y representa la probabilidad de los datos bajo cierta hipótesis.
* $P(B)$ se conoce como *probabilidad marginal*, y es una constante normalizadora que representa la probabilidad de los datos bajo cualquier hipótesis, o $P(A∪B)$. En otras palabras, puede calcularse como $P(A) P(B|A) + P(-A) P(B|-A)$ donde -A es No A (el escenario alternativo de la hipótesis).
* $P(A|B)$ se conoce como *probabilidad posterior*, y es la probabilidad que queremos calcular después de haber visto los datos.

**Importante:** Es importante destacar aquí que la hipótesis está representada por A y los datos observados por B. Un error común y un punto de confusión es pensar que A y B representan dos hipótesis diferentes, lo que no es el caso. Las dos hipótesis opuestas están representadas por A y -A.


## El Teorema de Bayes (Toni)

Con la probabilidad condicional utilizamos información adicional para saber la probabilidad de los efectos dados las causas, pero, muchas veces, lo que queremos saber es la probabilidad de las causas dadas los efectos. Por ejemplo, sabemos cuál es el porcentaje de pacientes con gripe que tiene fiebre, pero lo que realmente queremos saber es la probabilidad de que un paciente con fiebre tenga gripe. El teorema de Bayes nos permite ir de uno a otro con suma facilidad.

El Teorema de Bayes nos permite ir de una probabilidad condicional a su alterna con una matemática relativamente simple, porque, recordemos:

                p(A/B) ≠ p(B/A)


Bayes le dió un par de vueltas a la ecuación...

    Probabilidad condicional: p(A|B) = p(A∩B)/p(B) ➝ p(A∩B) = p(A|B) * p(B)
    Intersección A,B: p(A∩B) = p(A|B) * p(B) 
    Intersección B,A: p(B∩A) = p(B|A) * p(A)

                    OJO!

    Intersección A,B = Intersección B,A

                     ↓

    p(A|B) * p(B) = p(B|A) * p(A) ➞ p(B|A) = (p(A|B) * p(B)) / p(A) TEOREMA DE BAYES

Ejemplo Bayes:
Vamos a pensar en la aplicación de la inferencia bayesiana con un simple ejemplo del diagnostico médico, uno de los campos dónde más éxito ha tenido. Supongamos que nos hicimos un estudio y nos ha dado positivo para una rara enfermedad que solo el 0.3 % de la población tiene. La tasa de efectividad de este estudio es del 99 %, es decir, que solo da falsos positivos en el 1 % de los casos. ¿Cuán probable es que realmente tengamos la enfermedad?.

En un principio, nos veríamos tentados a responder que hay un 99 % de probabilidad de que tengamos la enfermedad; pero en este caso nos estaríamos olvidando del concepto importante del a priori. Sabemos con anterioridad que la enfermedad es extremadamente rara (solo el 0.3 % la tiene); si incluimos esta información previa en nuestro cálculo de probabilidad y aplicamos el teorema de Bayes podemos llegar a una conclusión totalmente distinta.

```python
# El test tiene una probabilidad del 0.99 de acertar (probabilidad de positivo en gente afectada) p(A|B)
# Hay una probabilidad del 0.003 de que estemos afectados (información adicional, 'a priori') p(B)
# Hay una probabilidad del 0.1 que el test haya fallado
# La probabilidad de positivos es la unión de los positivos a gente enferma y de los positivos a gente no enferma;
        # p(A) = p(A|B) * p(B) +  ~p(A|B) * ~p(B)
# Fórmula Bayes p(B|A) = (p(A|B) * p(B)) / p(A)
# Queremos saber la probabilidad condicional de que estemos afectados si nos da positivo p(B|A)
prob_b = 0.003
prob_cond_de_positivo_a_afectados = 0.99
prob_a = 0.99 * 0.003 + 0.01 * 0.997

prob_estemos_afectados = (prob_cond_de_positivo_a_afectados * prob_b) / prob_a 
prob_estemos_afectados
```

## Ejercicio Teorema de Bayes

OJO! Para calcular la probabilidad posterior según el teorema de bayes con múltiples eventos necesitamos calcular la probabilidad marginal de una forma distinta.

$P({E}_{1}|B) = \frac{P(B|{E}_{1})P({E}_{1})}{P(B|{E}_{1})P({E}_{1})+P(B|{E}_{2})P({E}_{2})+...+P(B|{E}_{n})P({E}_{n})}$

Podemos definir una función en Python que acepte una lista de antecedentes y sus correspondientes probabilidades. Nuestra función calculará el marginal (marg) y luego utilizará el Teorema de Bayes para calcular las probabilidades posteriores para cada una de las hipótesis (post).

```python
import numpy as np

def bayes_rule(priors, likelihoods):
    marg = sum(np.multiply(priors, likelihoods))
    post = np.divide(np.multiply(priors, likelihoods), marg)
    return post
```

Aplicación del Teorema de Bayes
Ahora que tenemos nuestra función, propongamos un escenario y usemos nuestra función para ayudarnos a resolver las probabilidades posteriores.

Supongamos que tenemos 3 frascos con 100 canicas en cada uno.

* El frasco 1 tiene 40 canicas azules, 30 canicas rojas y 30 canicas verdes.
* El frasco 2 tiene 60 canicas azules, 20 canicas rojas y 20 canicas verdes.
* El frasco 3 tiene 10 canicas azules, 30 canicas rojas y 70 canicas verdes.

Escoges una canica de uno de los frascos al azar, y la canica es verde. ¿Cuáles son las probabilidades de que la canica provenga de cada uno de los tres frascos?

Para responder a esta pregunta, necesitaríamos saber cuáles eran nuestras probabilidades previas y nuestras estimaciones. Para las previas, podemos asumir que la probabilidad de que saquemos la canica de cualquier frasco es igual de probable, o 1/3. Las estimaciones en este caso serían sólo los porcentajes de canicas verdes en cada frasco. Podemos introducir estos valores a nuestra función `bayes_rule`, y devolverá las siguientes salidas.

```python
np.multiply(priors, likelihoods)
```

```python
priors = [1/3, 1/3, 1/3]
likelihoods = [0.3, 0.2, 0.7]

bayes_rule(priors, likelihoods)
```

De estos resultados, podemos ver que hay un 25% de probabilidad de que la canica verde que dibujamos provenga del frasco 1, un 16.67% de probabilidad de que provenga del frasco 2, y un 58.3% de probabilidad de que provenga del frasco 3.

Supongamos que hubiera elegido una canica azul en lugar de una verde. Sólo tendríamos que actualizar nuestras estimaciones para tener en cuenta la proporción de cada frasco cuyas canicas eran azules.

```python
likelihoods = [0.4, 0.6, 0.1]

bayes_rule(priors, likelihoods)
```

Hay un 36.36% de probabilidad de que la canica verde que elegimos provenga del frasco 1, un 54.54% de probabilidad de que provenga del frasco 2, y sólo un 9% de probabilidad de que provenga del frasco 3.

## Análisis de datos desde una perspectiva bayesiana
El ejemplo anterior era simple con el objetivo de demostrar cómo funciona el Teorema de Bayes. Cuando se hace el Análisis de Datos Bayesiano en el mundo real, los previos y posteriores son típicamente distribuciones de probabilidad. Tenemos algunos datos, y la forma de inferir de lo previo a lo posterior es un modelo generativo (uno que acepta parámetros y genera datos a partir de estos).

Veamos un ejemplo para hacernos una idea de cómo funciona esto. Supongamos que acabamos de lanzar una tienda en línea que vende spinners, y estamos tratando de estimar qué porcentaje de visitantes a nuestra tienda en línea hará una compra. Podemos establecer esto como un problema bayesiano y llegar a una distribución de probabilidad de la siguiente manera.

Antes de ver cualquier dato de tráfico o compra no tendremos idea de qué esperar, así que empezaremos con la suposición de que todos los porcentajes son igualmente probables. Utilizaremos una distribución uniforme para representar el hecho de que cada tarifa tiene la misma oportunidad de ser la tarifa real de compra. Para establecer esto en Python, vamos a usar `Numpy` para generar una distribución uniforme a partir de 100.000 extracciones aleatorias de números entre 0 y 1.

```python
%matplotlib inline
import pandas as pd

n_draws = 100000
prior = pd.Series(np.random.uniform(0, 1, size=n_draws))
```

Esta es nuestra distribución previa, si la dibujamos nos devuelve el siguiente histograma:

```python
prior.hist();
```

Ahora es el momento de observar algunos datos y ver cómo afecta esto a nuestras estimaciones. Supongamos que el primer día 50 personas visitaron nuestro sitio, y 10 de ellas hicieron una compra. Vamos a registrar el número de compras en una variable llamada `observed`.

```python
observed = 10
```

Ahora vamos a crear un modelo generativo que extraerá aleatoriamente de nuestra distribución uniforme anterior la llegada de 50 personas simuladas a nuestro sitio web una gran cantidad de veces, y verá cuántas veces obtenemos un resultado que está en línea con el resultado que hemos observado. Para hacer esto en Python, vamos a definir una función generative_model que acepte un parámetro de probabilidad previa y luego realice 50 sorteos binomiales usando esa probabilidad.

```python
def generative_model(param):
    result = np.random.binomial(50, param)
    return result
```

Vamos a crear una lista vacía que contendrá nuestros resultados simulados (sim_data) y luego los rellenaremos añadiendo los resultados cuando cada valor de nuestra distribución previa esté incluido en nuestro modelo generativo.

```python
pd.DataFrame(sim_data).hist()
```

```python
sim_data = list()

for p in prior:
    sim_data.append(generative_model(p))
```

Finalmente, vamos a llegar a nuestra distribución posterior seleccionando sólo los valores de nuestra distribución previa que generaron instancias en las que el resultado simulado coincidió con nuestro resultado observado de 10 compras.

```python
posterior = prior[list(map(lambda x: x == observed, sim_data))]
```

A continuación se muestra cómo es nuestra distribución posterior.

```python
posterior.hist();
```

Podemos ver con sólo mirar esto que la probabilidad actualizada de que alguien haga una compra en nuestra tienda en línea es probablemente entre el 10% y el 30% y lo más probable es que sea entre el 17% y el 20%.

También podemos calcular algunas estadísticas para nuestra distribución posterior además de visualizarla. Esto nos mostrará la media, el rango y la desviación estándar de nuestra distribución posterior además de los cuartiles.

```python
posterior.describe()
```

Podemos ver que tanto la media como la mediana son aproximadamente del 21%. Si quisiéramos, también podríamos calcular un intervalo de confianza del 90% para esta distribución de la siguiente manera.

```python
print(posterior.quantile(.025), '|', posterior.quantile(.975))
```

Finalmente, también podríamos redondear los valores en la distribución posterior a porcentajes enteros y calcular el más verosímil junto con su probabilidad. Esto se conoce como la estimación de máxima verosimilitud, que nos informa de la moda dentro de los valores de nuestra distribución y la probabilidad de que ocurra ese valor, esta probabilidad la podemos calcular de la siguiente forma:

$\frac{n\ cases\ P(.21)}{n\ total\ cases} = \frac{151}{1899} = 0.0795$

```python
rounded = posterior.round(2)
mode = rounded.mode()[0]
probability = list(rounded).count(mode)/len(rounded)
print('Maximum Likelihood Estimate: ', mode, '|',probability)
```

Esto nos dice que, dados los datos que tenemos hasta ahora, es muy probable que el 21% de los visitantes de nuestra tienda online realicen una compra.

A lo largo de este experimento hemos pasado de una distribución a priori a una distribución posterior, ahora vamos a explicar a través de notación cómo hemos ido de una a otra. Para empezar volvamos a mostrar ambas distribuciones:

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12,5))
axes[0].hist(prior)
axes[1].hist(posterior, color = "green")
```

Tomemos el 21% para llevar a cabo el ejemplo de forma teórica. Si tomamos el 21% de la distribución a priori este suceso va a tener cierta probabilidad $P(0.21)$ de obtener un valor para el parámetro del modelo generativo.

El siguiente paso será simular datos utilizando este parámetro, y quedarnos con aquellos datos simulados que coincidan con los datos que hemos observado de la realidad. En nuestro ejemplo anterior, realizabamos una simulación en la que obteníamos el número de éxitos sobre 50 ensayos respecto a la compra dentro de nuestra tienda online. Notaremos esta probabilidad como $P(10|0.21)$. De esta forma, ya disponemos del numerador para la ecuación de Bayes:

$P(0.21)P(10|0.21)$

Ya solo nos faltaría dividir esto por la suma total de las probabilidades de la generación de datos con todos los valores posibles para los parámetros. Esto nos devolverá la probabilidad para una tasa de compra del 21% dados los datos recogidos.

$P(0.21|10) = \frac{P(0.21)P(10|0.21)}{\sum{P(\theta)P(10|\theta)}}$

Este procedimiento podemos utilizarlo para todas las posibles tasas de conversión a compra de nuestra web para obtener la probabilidad a posteriori para cada una de ellas.

## Más allá del ejemplo

Imaginemos que queremos comparar un diseño web A (el que teníamos en nuestro ecommerce en el experimento anterior) y otro diseño web B. Al recopilar información de los usuarios que han llegado a cada uno de las webs observamos que:
* A la web A llegan 50 usuarios y compran 10
* A la web B llegan 50 usuarios y compran 25

Si queremos comprobar las diferencias en las probabilidades de compra para cada una de las webs mediante un método bayesiano debemos replicar nuestro experimento anterior con estos nuevos datos, simplemente creando una nueva variable `observed_b` para los datos del test B:

```python
observed_b = 25
```

```python
posterior_b = prior[list(map(lambda x: x == observed_b, sim_data))]
```

```python
posterior.hist();
posterior_b.hist();
```

```python
print(posterior.quantile(.025), '|', posterior.quantile(.975))
print(posterior_b.quantile(.025), '|', posterior_b.quantile(.975))
```

```python
rounded = posterior.round(2)
rounded_b = posterior_b.round(2)
mode = rounded.mode()[0]
mode_b = rounded_b.mode()[0]
probability = list(rounded).count(mode)/len(rounded)
probability_b = list(rounded_b).count(mode_b)/len(rounded_b)
print('Maximum Likelihood Estimate Web A: ', mode, '|',probability)
print('Maximum Likelihood Estimate Web B: ', mode_b, '|',probability_b)
```

Esto está muy bien para entender los datos de una forma descriptiva, pero para testear que realmente existe una diferencia entre estas dos situaciones experimentales debemos calcular las diferencias entre las dos distribuciones posteriores:

```python
post_b = posterior_b.reset_index()
post_b.columns = ['index','prob']
post_a = posterior.reset_index()
post_a.columns = ['index','prob']
```

```python
posterior_diff = post_b["prob"] - post_a["prob"]
posterior_diff.hist()
```

En el histograma podemos observar que prácticamente no existen valores por debajo de 0. Vamos a confirmar al 95% de confianza que la web B tiene más posibilidades de éxito en la compra que la web A:

```python
print(posterior_diff.quantile(.025), '|', posterior_diff.quantile(.975))
```

Para obtener un valor probabilístico lo que debemos hacer es sumar el número de veces que ocurre que la probabilidad de éxito de la web B tiene mayor éxito que la web A y dividirlo por el número total de casos

```python
sum(posterior_diff > 0)/posterior_diff.count()
```

Podemos asumir que en el 99.9% de los casos la web B tendrá mayor éxito que la web A

Otra cosa que podemos hacer con Bayes, y que no es posible realizar desde un punto de vista clásico, es utilizar información experta a priori acerca de la distribución de los casos. Imaginemos que nos dicen desde la empresa que lo más normal no es lo que ha pasado en nuestro primer experimento, sino que usualmente los ratios de conversión a venta están entre el 5% y el 10%. Podemos utilizar una distribución a priori que tenga en cuenta estos ratios para aproximarnos más a la realidad. Para ello utilizaremos una distribución `beta` en lugar de una uniforme, esto nos permitirá manipular la forma de la distribución inicial:

```python
prior_2 = pd.Series(np.random.beta(2.5, 35, size=n_draws))
```

```python
prior_2.hist()
```

```python
posterior_2 = prior_2[list(map(lambda x: x == observed_b, sim_data))]
```

```python
posterior_2.hist()
```

```python
print(posterior_2.quantile(.025), '|', posterior_2.quantile(.975))
```
